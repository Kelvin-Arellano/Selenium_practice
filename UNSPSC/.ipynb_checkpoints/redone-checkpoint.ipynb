```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.firefox.webdriver import Options
import pandas as pd
from collections import Counter
from time import sleep
from random import choice
import re
import pickle
```

```python
df = pd.read_csv('Copy of KelvinUNSPSCProject.xlsx - All In one.csv')
```

```python
non_null_com = df[~df['Commodity/Level 4'].isnull()]
non_null_com = non_null_com[non_null_com['UNSPSC'].isnull()]
list_of_names = non_null_com['Commodity/Level 4'].to_list()
list_of_names_2 = []
for i in list_of_names:
    if i not in list_of_names_2:
        list_of_names_2.append(i)
```

```python
search_engine_choice = ['duck', 'yahoo','aol']
driver =webdriver.Firefox(executable_path=GeckoDriverManager().install())
for query in list_of_names_2:
    choice_ = choice(search_engine_choice)
    if choice_ == 'duck':
        search_string = query.replace(',','').replace('&','').replace(' ',"+")
        search_string = search_string+'+unspsc'
        driver.get(f'https://duckduckgo.com/?q={search_string}&t=h_&ia=web')
        sleep(2)
        driver.refresh()
        results_1 = driver.find_element(By.XPATH,'//*[@id="links"]').text
        results_numbers = re.findall("(\d{8})", results_1)
        unspsc_dict[f'{search_string}'] = results_numbers
    elif choice_ == 'yahoo':
        search_string = query.replace(',','').replace('&','').replace(' ',"+")
        search_string = search_string+'+unspsc'
        driver.get('https://search.yahoo.com/search;_ylt=AwrJ7Fr6Nv9hI98ArHtDDWVH;_ylc=X1MDMTE5NzgwNDg2NwRfcgMyBG'+
                   'ZyAwRmcjIDcDpzLHY6c2ZwLG06c2ItdG9wBGdwcmlkA2h2MkM5SXpjUzVHcTFYVndUb0F1bkEEbl9yc2x0AzAEbl9zdWdnA'+
                   'zAEb3JpZ2luA3NlYXJjaC55YWhvby5jb20EcG9zAzAEcHFzdHIDBHBxc3RybAMwBHFzdHJsAzE4BHF1ZXJ5A3dvb2QlMjBzY'+
                   f'3Jld3MlMjB1bnNwc2MEdF9zdG1wAzE2NDQxMTU3MTQ-?p={search_string}&fr=sfp&fr2=p%3As%2Cv%3Asfp%2Cm%3Asb-'+
                   'top&iscqry=')
        sleep(2)
        driver.refresh()
        results_1 = driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div/div/div[1]/div/div/div/div/ol').text
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        page_2 = driver.find_elements(By.XPATH,'/html/body/div[1]/div[3]/div/div/div[1]/div/ol/li[1]/div/div/div/a[1]')
        if len(page_2) > 0:
            page_2[0].click()
            sleep(2)
            results_2 = driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div/div/div[1]/div/div/div/div/ol').text
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        else:
            continue
        page_3 = driver.find_elements(By.XPATH,'/html/body/div[1]/div[3]/div/div/div[1]/div/ol/li[1]/div/div/a[2]')
        if len(page_3) > 0:
            page_3[0].click()
            results_3 = driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div/div/div[1]/div/div/div/div/ol').text
        else:
            continue
        results = results_1+results_2+results_3
        results_numbers = re.findall("(\d{8})", results)
        unspsc_dict[f'{search_string}'] = results_numbers
    elif choice_ == 'aol':
        search_string = query.replace(',','').replace('&','').replace(' ','+')
        search_string = search_string+'+unspsc'
        driver.get(f'https://search.aol.com/aol/search;_ylt=A2KLfSJ_SwFirBgAq1JoCWVH;_ylc=X1MDMTE5NzgwMzg4MARfcgMyBG'+
                   'ZyAwRmcjIDc2ItdG9wLXNlYXJjaARncHJpZANLRWVqWlB6SlRXcURJcElMUXNuRHNBBG5fcnNsdAMwBG5fc3VnZwMxMARvcm'+
                   'lnaW4Dc2VhcmNoLmFvbC5jb20EcG9zAzAEcHFzdHIDBHBxc3RybAMwBHFzdHJsAzExBHF1ZXJ5A3dvb2QlMjBzY3Jld3MEdF9z'+
                   f'dG1wAzE2NDQyNTIwNDA-?q={search_string}&v_t=na&s_it=sb-home&iscqry=&fr=&fr2=sb-top-search')
        sleep(1)
        driver.refresh()
        results_1 = driver.find_element(By.XPATH,'//*[@id="left"]').text
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        page_2 = driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div/div/div[1]/div/ol[2]/li/div/div/div/a[1]').click()
        sleep(1)
        results_2 = driver.find_element(By.XPATH,'//*[@id="left"]').text
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        page_3 = driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div/div/div[1]/div/ol[2]/li/div/div/div/a[2]').click()
        sleep(1)
        results_3 = driver.find_element(By.XPATH,'//*[@id="left"]').text
        results = results_1+results_2+results_3
        results_numbers = re.findall("(\d{8})", results)
        unspsc_dict[f'{search_string}'] = results_numbers
driver.quit()
```

```python
unspsc_file = open('unspsc_num.txt', 'wb')
pickle.dump(unspsc_dict, unspsc_file)
unspsc_file.close()
```
